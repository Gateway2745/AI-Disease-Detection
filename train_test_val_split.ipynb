```python
#first create splits for train and test_val
from sklearn.model_selection import ShuffleSplit

x_train_indices = []
x_test_val_indices = []

ss = ShuffleSplit(n_splits=5, test_size=0.20, random_state=0)
for train_index, test_index in ss.split(all_images):
    x_train_indices.append(train_index)
    x_test_val_indices.append(test_index)
```

```python
print('***** TRAIN , TEST_VAL SET DETAILS *****')
print(f'size of training set - {len(x_train_indices[0])}')
print(f'size of combined test and val set - {len(x_test_val_indices[0])}')
```

```python
#split test_val set to 50% test and 50% val set
import numpy as np

x_test_indices = []
x_val_indices = []

for test_val in x_test_val_indices:
    temp = np.random.permutation(test_val)
    x_test_indices.append(temp[:75])
    x_val_indices.append(temp[75:])
```

```python
print('***** TEST , VAL SET DETAILS *****')
print(f'size of test set - {len(x_test_indices[0])}')
print(f'size of val set - {len(x_val_indices[0])}')
```
