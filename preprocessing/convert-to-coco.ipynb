```python
import os

dir_name = '/home/rohit-ubuntu18_04/workdir/train-set-1'
img_dir = os.path.join(dir_name, 'images')
ann_dir = os.path.join(dir_name, 'annotations')
```

```python
image_list = sorted(os.listdir(img_dir))
ann_list = sorted(os.listdir(ann_dir))
```

```python
import json
import cv2
```

```python
images = []
annotations = []
```

```python
for img,ann in zip(image_list, ann_list):
    image = cv2.imread(os.path.join(img_dir, img))
    img_dict = {
        'file_name':img,
        'height':image.shape[0],
        'width':image.shape[1],
        'id': img[:-4]
    }
    
    images.append(img_dict)
    
    with open(os.path.join(ann_dir,ann), 'r') as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip().split(',')
            ann_dict = {
                'image_id': img[:-4],
                'bbox': line[:4],
                'category_id': line[-1],
                'id': ann[:-4]
            }
            annotations.append(ann_dict)

categories = [
    {
        'id':0,
        'name':'healthy_leaf'
    },
    {
        'id':1,
        'name':'diseased_leaf'
    },
    {
        'id':2,
        'name':'healthy_ear'
    },
    {
        'id':3,
        'name':'diseased_ear'
    }
]
```

```python
coco_dict = {
    'images':images,
    'annotations':annotations,
    'categories':categories
}
```

```python
coco_save_file = os.path.join(dir_name, '../train-set-1.json')
```

```python
with open(coco_save_file, 'w') as f:
    json.dump(coco_dict, f) 
```
